```
import pandas as pd
import numpy as np
```

```
train_df=pd.read_csv('datasets/fashion-mnist_train.csv')
test=df=pd.read_csv('datasets/fashion-mnist_test.csv')
```

```
train_df
```

```
train_label_df=train_df['label']
train_img_df=train_df.drop(['label'],axis=1)
train_img_matrix=train_df.drop(['label'],axis=1).to_numpy()
train_img_matrix=train_img_matrix.reshape(train_img_matrix.shape[0],28,28,1)
train_img_matrix=train_img_matrix/255.0
```

```
from sklearn.model_selection import train_test_split
```

```
X_train,X_test,y_train,y_test=train_test_split(train_img_matrix,train_label_df,random_state=42)
```

```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```
inputs=keras.Input(shape=(28,28,1))
x=layers.Conv2D(filters=32,kernel_size=(3,3),padding='same')(inputs)
x=layers.MaxPool2D()(x)
x=layers.Conv2D(filters=64,kernel_size=(3,3),padding='same')(x)
x=layers.MaxPool2D()(x)
x=layers.Flatten()(x)
x=layers.Dense(10)(x)
outputs=layers.Softmax()(x)
```

```
model=keras.Model(inputs=inputs,outputs=outputs)
```

```
model.summary()
```

```
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```
hist=model.fit(X_train,y_train,validation_split=0.1,epochs=5)
```

```
model.evaluate(X_test,y_test)
```
